```python
import numpy as np
import matplotlib.pyplot as plt 

def f(x, y):
    return x + np.cos(y / np.sqrt(3))

a = 1.2  # ліва межа відрізку
b = 2.2  # права межа відрізку
h = 0.2  # крок
y0 = 2.1  # початкова умова

n = int((b - a) / h)  # кількість кроків

x = np.array([i * h for i in range(n + 1)])  # задаємо x генератором списків

y = np.empty(n + 1) 
y[0] = y0

for i in range(n):
    y[i + 1] = y[i] + f(x[i], y[i]) * h

y_rounded = np.round_(y, 4) 
print("x=", x, "\ny=", y_rounded) 

plt.plot(x, y, "o-", color="red")  # Connect points with lines and use red color
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Метод Ейлера") 
plt.legend(["точки", "x + cos(y/√3)"])  # Correct the legend text
plt.grid() 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

def f(x, y):
    return x + np.sin(y / np.sqrt(2))

a = 0.8  # ліва межа відрізку
b = 1.8  # права межа відрізку
h = 0.2  # крок
y0 = 1.3  # початкова умова
n = int((b - a) / h)  # кількість кроків
x = np.array([i * h for i in range(n + 1)])  # задаємо x генератором списків

y = np.empty(n + 1) 
y[0] = y0

for i in range(n): 
    y[i + 1] = y[i] + (f(x[i], y[i]) + f(x[i + 1], y[i] + f(x[i], y[i]) * h)) * h / 2

y_rounded = np.round_(y, 4)

print("x=", x, "\ny=", y_rounded) 
plt.plot(x, y, "o-", color="red")  # Connect points with lines and use red color
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Метод Ейлера-Коші") 
plt.legend(["точки", "x + sin(y/√2)"])  # Update the legend text
plt.grid() 
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dx
def model(y,x):
    x + np.sin(y / np.sqrt(2))
    return x + np.sin(y / np.sqrt(2))

# initial condition
y0 = 1

# x points
x = np.array ([i*0.2 for i in range(0, 6 )])

# solve ODE
y = odeint(model,y0,x)
print('x=', x)
print('y=',y)

# plot results
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Розв’язання диф. рівняння')
plt.grid() 
plt.show()
```
